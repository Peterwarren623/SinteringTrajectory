```python
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
from sklearn.metrics import r2_score
```

```python
# density definitions
# density will climb from 0.64 to ideally 1 (but always slightly less) (somewhere between 0.85 and 1)
# we will take density from 0.64 to 0.95 (1000 points in between)
# we can add in some gaussian noise here as GN

G0 = 5
k = 10

t = np.arange(1440, dtype=np.float64)

G = ((G0**3)+k*t)**(1/3)

plt.plot(t, G, label="Density")
plt.xlabel('Time')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
theta = 0.6
eps = ((theta*G0)/G)**2

f = 1-eps
```

```python
plt.plot(t, f, label="Density")
plt.xlabel('Time')
plt.ylabel('Density')
plt.legend()
plt.show()

plt.plot(t, G, label="Density")
plt.xlabel('Time')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
a = 3.3
b = 3.6

S = a-b*f
```

```python
plt.plot(f, G, label="Density")
plt.xlabel('Density')
plt.ylabel('Grain Size')
plt.legend()
plt.show()

plt.plot(f, S, label="Density")
plt.xlabel('Density')
plt.ylabel('Normalized Surface area')
plt.legend()
plt.show()
```
